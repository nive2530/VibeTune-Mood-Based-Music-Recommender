```python
# Converting the strings to lower
df['lyrics'] = df['lyrics'].astype(str).str.lower()
```

```python
# Replacing the new line to space
df['lyrics'] = df['lyrics'].apply(lambda x : x.replace('\n', ' ').split(' '))
```

```python
# Getting correct Lyrics in neat format
def lyrics_extraction(lyrics_list):
    lyr = []
    for i in lyrics_list:
        match = re.match(r"(\W+)([a-zA-Z]+)(\W+)", i) 
        if match:
            lyr.append(match.group(2)) 
    lyr = set(lyr)
    tags = ''
    for word in lyr:
        tags += word+' '
    return tags

    
```

#### Derving Lyric Tags from lyrics column

```python
df['lyric_tags'] = df['lyrics'].apply(lambda x : lyrics_extraction(x))
```

```python
df.head()
```

```python
def get_mood_label(row):
    energy = row['energy']
    valence = row['valence']
    tempo = row['tempo']
    danceability = row['danceability']
    acousticness = row['acousticness']
    liveness = row['liveness']

    if energy >= 0.5 and valence >= 0.6 and tempo >= 0.5 and danceability >= 0.5:
        return ["Happy","joyful", "upbeat", "bright", "carefree", "cheerful", "bubbly", "playful", "lively"]
    elif energy <= 0.4 and valence <= 0.3 and tempo <= 0.4:
        return ["gloomy", "heartbroken", "tearful", "lonely", "sorrowful", "nostalgic", "melancholy", "mellow", "Sad"]
    elif energy <= 0.5 and 0.3 <= valence <= 0.6 and tempo <= 0.6 and acousticness >= 0.4:
        return ["Chill", "calm", "soothing", "laid-back", "breezy", "ambient", "tranquil", "soft", "relaxing"]
    elif energy >= 0.6 and valence <= 0.4 and tempo >= 0.6:
        return ["intense", "fierce", "raw", "rebellious", "furious", "aggressive", "loud", "Angry"]
    elif 0.4 <= energy <= 0.8 and 0.4 <= valence <= 0.8 and 0.3 <= tempo <= 0.7:
        return  ["passionate", "affectionate", "dreamy", "tender", "loving", "sentimental", "heartfelt", "longing", "Romantic"]
    elif energy <= 0.4 and valence <= 0.2 and acousticness >= 0.5:
        return ["eerie", "haunting", "mysterious", "brooding", "suspenseful", "cinematic", "shadowy", "Dark"]
    elif energy >= 0.6 and valence >= 0.5 and tempo >= 0.6 and danceability >= 0.5:
        return ["Energetic", "pulsing", "bouncing", "fast-paced", "groovy", "rhythmic", "hype", "club", "vibe"]
    else:
        return ['Mixed']
```

```python
df['mood'] = df.apply(get_mood_label, axis=1)
```

### Cleaning the tags

```python
# Cleaning the tags
def clean_tags(text):
    tokenize = word_tokenize(text)
    filtered_words = [word for word in tokenize if word not in stopwords.words('english')]
    return ' '.join(filtered_words)
```

```python
def create_corpus(tags):
    corpus = ""
    for tag in tags:
        corpus += tag + ' '
    return corpus.strip()
        
```

```python
df['corpus'] = df['tags_tokenized'].apply(lambda x : create_corpus(x))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
def recommend_by_embedding_similarity(input_text, df, top_n=10):
    # Combine tags + lyrics to form the corpus
    total_corpus = df['corpus'].tolist()
    corpus.insert(0, input_text)  # Add input text as the first entry

    # TF-IDF Vectorization
    tfidf = TfidfVectorizer()
    tfidf_matrix = tfidf.fit_transform(corpus).toarray()

    # Compute cosine similarity: input vs. all others
    input_vector = tfidf_matrix[0].reshape(1, -1)
    corpus_vectors = tfidf_matrix[1:]

    similarities = cosine_similarity(input_vector, corpus_vectors)[0]
    top_indices = similarities.argsort()[-top_n:][::-1]

    # Collect top results
    recommendations = df.iloc[top_indices][['track_name', 'artist_name', 'spotify_url']]
    recommendations['similarity_score'] = similarities[top_indices]
    return recommendations

```

```python
corpus = df['track_name'].str.lower().str.split().tolist()
model = Word2Vec(corpus, vector_size=300, window=5, min_count=1, workers=4)
```

```python
def song_recommendation(song_name, model, df):
    try:
        # Get the vector for the input song
        input_vector = model.wv[song_name.lower()]
    except KeyError:
        return f"'{song_name}' not found in the vocabulary."

    # Create a matrix of vectors for all songs in df
    all_songs = df['track_name'].str.lower().tolist()
    
    # Filter only those that exist in the Word2Vec vocabulary
    valid_songs = [song for song in all_songs if song in model.wv]
    vectors = [model.wv[song] for song in valid_songs]

    # Compute cosine similarity between input song and all valid songs
    similarities = cosine_similarity([input_vector], vectors)[0]

    # Pair songs with their similarity scores
    song_similarity = list(zip(valid_songs, similarities))

    # Sort by similarity (excluding the input song itself)
    song_similarity = sorted(song_similarity, key=lambda x: x[1], reverse=True)
    song_similarity = [song for song in song_similarity if song[0] != song_name.lower()]

    # Return top 6 similar songs
    recommended_songs = [song[0] for song in song_similarity[:6]]
    return recommended_songs
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def get_similar_tracks(input_text, model, df, top_n=5):
    input_tokens = input_text.lower().split()
    valid_tokens = [token for token in input_tokens if token in model.wv]

    if not valid_tokens:
        return "No valid words in input found in Word2Vec vocabulary."

    # Average vector for the input text
    input_vector = np.mean([model.wv[token] for token in valid_tokens], axis=0).reshape(1, -1)

    # Compute vectors for all valid songs
    song_vectors = []
    valid_rows = []

    for _, row in df.iterrows():
        track = row['track_tags'].lower()
        tokens = track.split()
        track_tokens = [token for token in tokens if token in model.wv]
        if not track_tokens:
            continue
        vector = np.mean([model.wv[token] for token in track_tokens], axis=0)
        song_vectors.append(vector)
        valid_rows.append(row)

    # Compute cosine similarity
    similarities = cosine_similarity(input_vector, song_vectors)[0]

    # Rank and return top N results
    ranked = sorted(zip(valid_rows, similarities), key=lambda x: x[1], reverse=True)
    top_similar = [
        {"track_name": row['track_name'], "spotify_url": row['spotify_url']}
        for row, _ in ranked[:top_n]
    ]

    return top_similar
```
